```python
# import thr necessary libraries
from google import genai
from google.genai import types 
import os
from dotenv import load_dotenv

## load the env date from folder
load_dotenv(dotenv_path=".env")

API_KEY = os.environ.get('GEMINI_API_KEY')

client = genai.Client(api_key=API_KEY)
response = client.models.generate_content(
    model='gemini-2.5-flash-lite',
    contents = "my name is ganesh , i am currently learning ai from stracth, generate a message to recruiter",

)
print(response.text)
```

## Giving instructions to AI how to answer


```python
## tell ai to behave like romantic poet
promt_1 = 'generate the text to make pizza'

print('='*40)

response = client.models.generate_content(
    model ='gemini-2.5-flash-lite',
    contents = promt_1,
    config= types.GenerateContentConfig(
        system_instruction = ' you are a romantic poet, let answe in a way that use flowers and more'
    )
)
print(response.text)
print('/n')
```

```python
## example 2 to tell ai to behave like professional
promt_1 = 'generate the text to make pizza'

print('='*80)

response = client.models.generate_content(
    model ='gemini-2.5-flash-lite',
    contents = promt_1,
    config= types.GenerateContentConfig(
        system_instruction = ' you are a professional, use professional tone to explain'
    )
)
print(response.text)
print('/n')
print('='*80)
```
